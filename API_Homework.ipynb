```python
## packages

import requests
import pandas as pd
import json
```

```python
# users data

url = "https://jsonplaceholder.typicode.com/users/"

r = requests.get(url)

r.json()
```

```python
#first dataframe

df_temp = pd.read_json(url)

df1 = df_temp.rename(columns = {"id" : "userId"})

df1

## renaming id column to be able to join through same column name

```

```python
## posts data

url2 = "https://jsonplaceholder.typicode.com/posts/"

r2 = requests.get(url2)

r2.json()
```

```python
#second dataframe (50 results)

df2 = pd.read_json(url2)

df50 = df2.sample(50)

df50
```

```python
#table merge on userID

df = pd.merge(df50, df1, on='userId')

df
```
